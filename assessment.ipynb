```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('netflix_data_clean.csv')
```

```python
plt.figure(figsize=(10, 6))
df.groupby('type')['imdb_score'].plot(kind='hist', alpha=0.5, legend=True)
plt.title('IMDb Score Distribution for Movie and Show')
plt.xlabel('IMDb Score')
plt.ylabel('Frequency')
plt.show()
```

```python
df_sorted = df.sort_values(by='imdb_score', ascending=False)

top_100_titles = df_sorted.head(100)

plt.figure(figsize=(12, 6))
sns.countplot(y='type', data=top_100_titles, palette='viridis')
plt.title('Distribution of Movies and Shows in Top 100 IMDb Scores')
plt.xlabel('Count')
plt.ylabel('Type')
plt.show()
```

```python
df_sorted = df.sort_values(by='imdb_score', ascending=False)

top_100_titles = df_sorted.head(100)

plt.figure(figsize=(12, 6))
sns.barplot(x='type', y='runtime', data=top_100_titles, errorbar=None, palette='viridis')
plt.title('Average Runtime for Shows and Movies in Top 100 IMDb Scores')
plt.xlabel('Type')
plt.ylabel('Average Runtime')
plt.show()
```

```python
average_runtime_per_year = df.groupby(['release_year', 'type'])['runtime'].mean().reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(x='release_year', y='runtime', hue='type', data=average_runtime_per_year, errorbar=None, palette='viridis')
plt.title('Average Runtime per Year for Movies and Shows')
plt.xlabel('Release Year')
plt.ylabel('Average Runtime')
plt.show()
```

```python
high_score_titles = df[df['imdb_score'] > 8]

titles_above_8_per_year = high_score_titles.groupby('release_year')['title'].count().reset_index()

plt.figure(figsize=(20, 6))
sns.barplot(x='release_year', y='title', data=titles_above_8_per_year, palette='viridis')
plt.title('Number of Titles with IMDb Score Above 8 per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```
